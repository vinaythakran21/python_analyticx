# 6. In customer table, create a new variable called “customer value segment” using customer value as follows. -  
#- High Value Segment - > 25000  
#- Medium Value Segment – Between 10000 and 25000  
#- Low Value Segment – less than or equal to 10000 

```python
cust['Customer_Value_Segment']=pd.cut(cust['Customer_Value'],bins=[0,10000,25000,5098715],labels=['Low Value Segment','Medium Value Segment','High Value Segment'])
cust.head()
```

# 7. Create variables “average revenue per trip” and “balance points” in the customer data set.  

```python
cust['Average_Revenue_Per_Trip']=cust['Customer_Value']/cust['Buy_Times']
cust['Balance_Points']=cust['Points_Earned']-cust['Points_Redeemed']
cust.head()
```

# 8. How many days between last purchase date and today?

```python
#converting first_date variable into a datetime variable
import datetime
cust['F_yr']=cust.First_Date.astype(str).str.slice(0,4)
cust['F_mn']=cust.First_Date.astype(str).str.slice(4,6)
cust['F_dy']=cust.First_Date.astype(str).str.slice(-2)
cust['F_date']=cust['F_yr']+'/'+cust['F_mn']+'/'+cust['F_dy']
cust.drop('F_yr',axis=1,inplace=True)
cust.drop('F_mn',axis=1,inplace=True)
cust.drop('F_dy',axis=1,inplace=True)
cust['First_Date']=pd.to_datetime(cust.F_date)
cust.drop('F_date',axis=1,inplace=True)
cust.head()
```

```python
#converting recent_date to datetime variable
cust['R_yr']=cust.Recent_Date.astype(str).str.slice(0,4)
cust['R_mn']=cust.Recent_Date.astype(str).str.slice(4,6)
cust['R_dy']=cust.Recent_Date.astype(str).str.slice(-2)
cust['R_date']=cust['R_yr']+'/'+cust['R_mn']+'/'+cust['R_dy']
cust.drop('R_yr',axis=1,inplace=True)
cust.drop('R_mn',axis=1,inplace=True)
cust.drop('R_dy',axis=1,inplace=True)
cust['Recent_Date']=pd.to_datetime(cust.R_date)
cust.drop('R_date',axis=1,inplace=True)
cust.head()
```

```python
#calculating number of days
d1=cust.sort_values(['First_Date'])
d1.set_index('First_Date',inplace=True)
d1.head()
```

```python
delta=d1.index[-1]-d1.index[0]
delta.days
```

# 9. Calculate percentage of sales by each last city, state and region.

```python
cust.groupby(['Last_region','Last_state','Last_city']).sum()['Customer_Value'].apply(lambda x: (x/cust.Customer_Value.sum())*100)
```
