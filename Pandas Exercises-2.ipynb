### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the wine dataset from this [address](https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data). 

### Step 3. Assign it to a variable called wine

```python
wine=pd.read_csv("https://archive.ics.uci.edu//ml//machine-learning-databases//wine//wine.data",header= None)
wine.head()
```

### Step 4. Delete the first, fourth, seventh, nineth, eleventh, thirteenth and fourteenth columns

```python
wine.drop([0,3,6,8,10,12,13],axis=1,inplace=True)
wine.head(10)
```

### Step 5. Assign the columns as below:

The attributes are:  
1) alcohol  
2) malic_acid  
3) alcalinity_of_ash  
4) magnesium  
5) flavanoids  
6) proanthocyanins  
7) hue 

```python
wine=wine.rename(columns={1:'alcohol',2:'malic_acid',4:'alcalinity_of_ash',5:'magnesium',7:'flavanoids',9:'proanthocyanins',11:'hue'})
wine.head(10)
```

### Step 6. Set the values of the first 3 rows from alcohol as NaN

```python
import numpy as np
wine.loc[0:2,'alcohol']=np.NaN
wine.head(10)
```

### Step 7. Now set the value of the rows 3 and 4 of magnesium as NaN

```python
wine.loc[3:4,'magnesium']=np.NaN
wine.head(10)
```

### Step 8. Fill the value of NaN with the number 10 in alcohol and 100 in magnesium

```python
wine.alcohol.fillna(10)
wine.magnesium.fillna(100)
```
